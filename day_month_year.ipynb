```python
#initializing varibales
#create a list for the date
#string varibales for date,day,month,and year

date_list=[]
date_string=''
day=''
month=''
year=''

#the function readoriginaldate will read in a string of characters
def readoriginaldate():
    
    #return date string value
    return date_string

#the function breakoriginaldate will break the original date into three pieces: a month, a day, and a two-digit year.
def breakoriginaldate():
    
    #print the original date
    print(f"{readoriginaldate()}\t\tis the original date")
    
    #break the date to the list
    #for example, month/day/two-digit-year will convert to the list [month,year,two-digit-year]
    break_date=readoriginaldate().split('/')
    
    #change the value of varibales
    global month,day,year
    month=str(break_date[0])
    day=str(break_date[1])
    year=str(break_date[2])
    
    #print the month, day, year
    print(f"{month} is the month\t{day} is the day\t{year} is the year")
    
    #return three string values
    return day,month,year

#the function printdate3ways which will receive three strings, the day, the month, and the year, as parameters.
def printdate3ways(day,month,year):
    
    #The first way will be the day, then a dash, then the month, then a dash, then the year. 
    european_print=f"{day}-{month}-{year} is the European way of printing"
        
    #The second way will be the English word for the month, the day, a comma, a space, and then the year written using 4 digits
    #create a dictionary for the month, then it can convert the number of month to english word
    month_dict = {'1': "January",'2': "February",'3':"March",
                '4': "April",'5': "May",'6':"June",
                '7': "July",'8': "August",'9':"September",
                '10': "October",'11': "November",'12':"December"}
    
    american_print=f"{month_dict.get(month)} {day},{int(year)+2000} is the American way of printing"
    
    #The third way will be the month written using two digits in all cases, a dash, the day written using two digits in all cases, a dash, and then the four-digit year.
    #create a dictionary have the key of original month, value for 2-digits-month
    month_dict_2_digit={'1': "01",'2': "02",'3':"03",
                '4': "04",'5': "05",'6':"06",
                '7': "07",'8': "08",'9':"09",
                '10': "10",'11': "11",'12':"12"}
    
    #if the day is one digit it should be convert to 2 digits
    if len(day)<2:
        day_full=list(day)
        day_full.insert(0,'0')
        day_2_digit=''
        for x in day_full:
            day_2_digit+=x
            
    #if the day is 2 digits, then it will keep the 2 digits
    else:
        day_2_digit=day
    
    full_way_print= f"{month_dict_2_digit.get(month)}-{day_2_digit}-{int(year)+2000} is the full way of printing"
    
    #print all 3 ways
    print(european_print,american_print,full_way_print,sep='\n')
    
#the main function calls all functions
def main():
    readoriginaldate()
    breakoriginaldate()
    printdate3ways(day,month,year)
```
